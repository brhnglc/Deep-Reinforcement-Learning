```python
import gym
env = gym.make('MountainCar-v0',render_mode="rgb_array")# mountaincar env'sini çağırır
#render mode ile statlerin çıktılarını görebilirz


#human yapıp #!pip install gym[classic_control] i indirisen local de video oynuyor oyunun
```

```python
import time 

num_steps = 1500
obs = env.reset()

for step in range(num_steps):

    action = env.action_space.sample()
    obs, reward, done, info = env.step(action)
    env.render()
    time.sleep(0.001)
    if done:
        env.reset()

env.close()
```

# env.
* **action_space** ->yapılabilecek eylem sayısını verir
* **observation_space** ->değişebilir yaygın kullanımı ekranın görüntüsü #vadi ortasında araba örnegi için observation_space = 2 hız ve konum
* **reset** -> env sıfılıra başlangıc stateine geçirir
* **step** -> aksiyonu input olarak alır ve ortama uygular sonuç olarak yeni env'ye geçilir 4 şey döndürür;
    *  *observation* = The observation of the state of the environment.
    *  *reward* = stepe verilen aksiyon sonucu alınan ödül
    *  *done* = true ise bölüm bitti tru ise env reset filan atabilirsin
    * *info* = ek bilgi sunar env'ye bağlı olarak mesela can sayısı gibi ya da debuggin için bilgi
* **action_space.sample()** -> rastgele bir action
